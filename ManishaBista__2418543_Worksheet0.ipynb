```python
def unit_converter():
    """
    Converts between different measurement units.

    Supported conversions:
    - Length: meters <-> feet
    - Weight: kilograms <-> pounds
    - Volume: liters <-> gallons

    Returns:
        float: Converted value
    """
    try:
        print("Select conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter choice (1/2/3): ")

        value = float(input("Enter value to convert: "))

        if choice == "1":
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            option = input("Choose option: ")

            if option == "1":
                result = value * 3.28084
                print("Result:", result, "ft")
            elif option == "2":
                result = value / 3.28084
                print("Result:", result, "m")

        elif choice == "2":
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            option = input("Choose option: ")

            if option == "1":
                result = value * 2.20462
                print("Result:", result, "lbs")
            elif option == "2":
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == "3":
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            option = input("Choose option: ")

            if option == "1":
                result = value * 0.264172
                print("Result:", result, "gal")
            elif option == "2":
                result = value / 0.264172
                print("Result:", result, "L")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid input. Please enter numeric value.")

unit_converter()
```

```python
def find_sum(numbers):
    """Returns the sum of numbers."""
    return sum(numbers)

def find_average(numbers):
    """Returns the average of numbers."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns the maximum value."""
    return max(numbers)

def find_min(numbers):
    """Returns the minimum value."""
    return min(numbers)


try:
    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter choice: ")

    nums = list(map(float, input("Enter numbers separated by space: ").split()))

    if len(nums) == 0:
        raise ValueError("Empty list")

    if choice == "1":
        print("Sum =", find_sum(nums))
    elif choice == "2":
        print("Average =", find_average(nums))
    elif choice == "3":
        print("Maximum =", find_max(nums))
    elif choice == "4":
        print("Minimum =", find_min(nums))
    else:
        print("Invalid option")

except ValueError:
    print("Invalid input.")
```

```python
def extract_every_other(lst):
    return lst[::2]

print(extract_every_other([1,2,3,4,5,6]))
```

```python
def reverse_list(lst):
    return lst[::-1]

print(reverse_list([1,2,3,4,5]))
```

```python
def remove_first_last(lst):
    return lst[1:-1]

print(remove_first_last([1,2,3,4,5]))
```

```python
def get_first_n(lst, n):
    return lst[:n]

print(get_first_n([1,2,3,4,5],3))
```

```python
def get_first_n(lst, n):
    return lst[:n]

print(get_first_n([1,2,3,4,5],3))
```

```python
def get_last_n(lst, n):
    return lst[-n:]

print(get_last_n([1,2,3,4,5],2))
```

```python
def reverse_skip(lst):
    return lst[-2::-2]

print(reverse_skip([1,2,3,4,5,6]))
```

```python
def flatten(lst):
    flat = []
    for sub in lst:
        for item in sub:
            flat.append(item)
    return flat

print(flatten([[1,2],[3,4],[5]]))
```

```python
def access_nested_element(lst, indices):
    result = lst
    for i in indices:
        result = result[i]
    return result

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
def find_max(lst):
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```
